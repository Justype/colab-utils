```python
#@markdown # 5️⃣ Quick Walk Through QC Reports

# Read files: output/*/fastqc_data.txt
import os
import glob
import pandas as pd

!cd output/ && unzip -nq *.zip  && cd ..
search_pattern = "output/*/fastqc_data.txt"
matching_files = glob.glob(search_pattern)

report = pd.DataFrame()

for file_path in matching_files:
  with open(file_path, 'r') as file:
    filename = ""
    for line in file:
      if line.startswith("Filename"):
        filename = line.strip().split("\t")[1]
      elif line.startswith("Total Sequences"):
        report.loc[filename, "N_Reads"] = line.strip().split("\t")[1]
      elif line.startswith("Sequences flagged as poor quality"):
        report.loc[filename, "N_Poor_Reads"] = line.strip().split("\t")[1]
      elif line.startswith("%GC"):
        report.loc[filename, "%GC"] = line.strip().split("\t")[1]
      elif line.startswith(">>Per base sequence quality"):
        report.loc[filename, "Base_quality"] = line.strip().split("\t")[1]
      elif line.startswith(">>Per sequence quality scores"):
        report.loc[filename, "Seq_quality"] = line.strip().split("\t")[1]
      elif line.startswith(">>Per base sequence content"):
        report.loc[filename, "Base_content"] = line.strip().split("\t")[1]
      elif line.startswith(">>Per sequence GC content"):
        report.loc[filename, "GC"] = line.strip().split("\t")[1]
      elif line.startswith(">>Sequence Length Distribution"):
        report.loc[filename, "Length_Distribution"] = line.strip().split("\t")[1]
      elif line.startswith(">>Adapter Content"):
        report.loc[filename, "Adapter"] = line.strip().split("\t")[1]
      elif line.startswith(">>Overrepresented sequences"):
        report.loc[filename, "Overrepresented"] = line.strip().split("\t")[1]

print(report.to_string())
```
